```python
from titanic_tools import TitanicData

td = TitanicData("../data/Titanic-Dataset.csv")
raw = td.load()
raw.head(), raw.columns.tolist(), raw.shape

```

```python
clean = td.clean_basic()
clean.head(), clean.dtypes, clean.isna().sum()

```

```python
from titanic_tools import TitanicAnalysis

ta = TitanicAnalysis(clean)
ta.survival_rate(), ta.survival_rate(by="sex"), ta.survival_rate(by=["sex","pclass"])

```

```python
import matplotlib.pyplot as plt

ta.plot_survival_by("sex")
plt.show()
```

```python
ta.plot_survival_by("pclass"); plt.show()
ta.plot_survival_by("embarked"); plt.show()
```
